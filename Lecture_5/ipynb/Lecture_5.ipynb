### Updating Pandas Series
<p class="grey"> Updating a Series will work as either UPDATE or APPEND </p>
<hr class="red">
* If the index is not present, we add a new value (APPEND BEHAVIOR)
* If the index is present, we may change the value (UPDATE BEHAVIOR)
* If duplicate indexes values are present all values will be updated

```python
dob_ser = pd.Series(["Rama","Adam","Meena","Jane"], index = [1978,1983,1985,1985])
print (dob_ser)
#wasn't present so we add it
dob_ser[1990] = "Nancy"
dob_ser
```

```python
#was present we updated
dob_ser[1978] = "Raj"
dob_ser
```

```python
dob_ser[1985] = 'Nemanja'
dob_ser
```

#### Adding values to a series via append()

```python
#name
dob_ser = pd.Series(["Rama","Adam","Meena","Jane"], index = [1978,1983,1985,1985], name = "DOB")
print (dob_ser)
dob_ser.append(pd.Series({"Alana":1980}))
dob_ser
#Notice that this looses the name of the original series so it's not really a true append
dob_ser2 = dob_ser.append(pd.Series({1980:"Alana"}))
dob_ser2
```

<center><h1 class = "blue"> Deleting values from a Series object </h1></center>
### This is NOT common, filtering is preferred. 

```python
dob_ser = pd.Series(["Rama","Adam","Meena","Jane"], index = [1978,1983,1985,1985], name = "DOB")
print (dob_ser)
del dob_ser[1983]
dob_ser
del dob_ser[1985]
dob_ser
```

```python
#we can create the isin method to create a mask and ~ to reverse the values of the mask
dob_ser = pd.Series(["Rama","Adam","Meena","Jane"], index = [1978,1983,1985,1985], name = "DOB")
print (dob_ser.isin(["Meena",'Jane']))
dob_ser2 = dob_ser[~dob_ser.isin(["Meena",'Jane'])]
dob_ser2
```

<center><h1 class = "blue"> Missing Data</h1></center>

```python
import pandas as pd
weather_m = pd.read_csv('weather_c.csv')
```

```python
weather_m
```

#### Do you have missing data?

```python
#Check if Value is missing or not
weather_m.isnull()
```

#### Drop all rows that have ANY missing values

```python
#Drop all rows that have ANY missing value
weather_m_dropped= weather_m.dropna()
weather_m_dropped
```

#### Drop rows only if any values are missing

```python
#specifying drop if ANY values are missing
weather_m_drop_ifall = weather_m.dropna(how = 'any')
weather_m_drop_ifall
```

#### Drop all rows that have have ALL values missing

```python
#specifying drop if ANY are missing or if ALL are missing
weather_m_drop_any = weather_m.dropna(how = 'all')
weather_m_drop_any
```

#### Setting a minimum treshold of values for keeping

```python
#asking for at least N values per row that are not NaN
weather_m
weather_m.dropna(thresh=3)
```

```python
weather_m
```

<center><h1 class = "blue"> Filling Missing Values With Data</h1></center>

#### Missing values can be replaced with any value

```python
### Filling missing values (Replacing with another value)
weather_m_zero = weather_m.fillna(0)
weather_m
```

#### Sometimes in numeric columns you may want to replace missing values with the mean of the column

```python
### A more useful thing to do may be to replace the vals with an Average
weather_mean = weather_m.copy()
weather_mean["Humidity Night"] = weather_mean["Humidity Night"].fillna(weather_mean["Humidity Night"].mean())
weather_mean
```

#### It is also possible to drop a whole column based on missing value status

```python
#you can also drop columns
weather_d = pd.read_csv('weather_d.csv')
print (weather_m)
weather_d.dropna(axis = 1, how ='all', inplace = True)
weather_d
```

<center><h1 class = "blue"> DataFrame Concatenation</h1></center>

```python
df1 = pd.DataFrame({'name':['Ann','Bob','Tijana'], 'color':['Blue','Red',np.nan]})
df2 = pd.DataFrame({'name':['Jane','Dana','Tihomir'], 'color':['Blue','Green','Green']}, index = [3,2,1])
print (df1)
print (df2)
```

#### This works but notice our index

```python
df3 = pd.concat([df1,df2])
df3
```

#### Reset a df index

```python
df3 = df3.reset_index(drop=True)
df3
```

#### You could reindex

```python
df3 = df3.reindex(range(0,len(df3.index)))
df3
```

#### How would you even know if your index has duplicates?

```python
df3 = pd.concat([df1,df2], verify_integrity = True)
```

#### We could just get started with a new index when we concatenate?

```python
df3 = pd.concat([df1,df2], ignore_index = True)
df3
```

#### We can also concatenate columns, be careful about columns with equal names.

```python
df3 = pd.concat([df1,df2], axis = 1)
df3['color']
```
