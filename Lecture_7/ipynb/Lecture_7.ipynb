### Using Scikit-learn

<hr class="red">

### <span class = "yellow"> fit() </span>  The fit method is used to train the model on training data

### <span class = "yellow"> predict() </span> The predict method is used to predict target variables of the new data

### Imporint and instantiating the classifier
<p class="grey"> Supervised Learning With Scikit </p>
<hr class="red">

```python
#IMPORT YOUR CLASSIFIER
from sklearn.neighbors import KNeighborsClassifier
#CREATE AN INSTANCE OF THE CLASSIFIER WITH k = 5
knn = KNeighborsClassifier(n_neighbors=5)
#FIT THE MODEL TO THE TRAINING DATA
knn.fit(iris['data'],iris['target'])
```

### Prediction of target variable for new data
<p class="grey"> Supervised Learning With Scikit </p>
<hr class="red">



```python
#let's create the new data
new = np.array([[6.8,3.2,5.7,2.4],
               [6.7,3.0,5.6,2.4],
               [5.0,3.3,1.5,0.2],
               [5.3,3.4,1.4,0.1]])

target_variables = knn.predict(new)
for i,j in zip(new,target_variables):
    print (i,iris['target_names'][j])
```

### How can we be sure that our model is performing well?
<p class="grey"> Model Performance Measurement </p>
<hr class="red">

* For classification tasks the metric we use is <span class = "yellow">Accuracy</span>
* <span class = "yellow">Accuracy = Number of correct predictions / Number of Data Points</span>
* Can we use some of the data we trained with to test the accuracy?  No that would be silly.
* Can we use new data to test the accuracy? No because we don't know the labels, this is our task.
* Instead we will split the data into two parts: training and testing.

### Splitting the data to training and testing parts
<p class="grey"> Scikit helps us with that </p>
<hr class="red">

```python
from sklearn.model_selection import train_test_split
X = iris.data
y = iris.target
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.3,random_state = 42, stratify = y)

#CREATE AN INSTANCE OF THE CLASSIFIER WITH k = 5
knn = KNeighborsClassifier(n_neighbors=7)
#FIT THE MODEL TO THE TRAINING DATA
knn.fit(X_train,y_train)

#LET'S MAKE SOME PREDICTIONS ON THE TEST DATA
y_predictions = knn.predict(X_test)
for i,j in zip(X_test,y_predictions):
    print (i,iris['target_names'][j])

#LET'S SEE IF OUR PREDICTIONS WERE ACCURATE
knn.score(X_test,y_test)
```

### Overfitting and underfitting
<p class="grey"> Picking K </p>
<hr class="red">

* Typically 1-10
* If you are categorizing a binary outcome don't pick k = 2
* Best to not pick a multiple of the number of categories
* Low K will cause overfitting and decrease bias but increase variance, 
* High K will cause underfitting and increase bias and lower variance, smooth decision boundary
